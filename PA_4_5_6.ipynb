```python
import pandas as pd
import numpy as np
import csv

with open("communities.data") as f:
    r = csv.reader(f)
    df = pd.DataFrame().from_records(r)
    

df= df.replace('?', np.nan) 

df.iloc[:, 3] = pd.to_numeric(df.iloc[:, 3], errors='coerce')
from sklearn.preprocessing import Imputer
fill_NaN =Imputer(missing_values=np.nan, strategy='mean', axis=0)
imputed_DF = pd.DataFrame(fill_NaN.fit_transform(df))
df = imputed_DF
df.to_excel("CandC_no_missing_Values.xlsx")
print("_________________________________________Answer to Question 4____________________________________")
```

```python
#--------------------------------------Linear Regression---------------------------------------------#

print("\n___________________________________Answer to Question 5______________________________________\n")
print("\n//////////All test and train sets have been stored in .csv files in the same folder////////////////////\n")
from sklearn.model_selection import train_test_split
X = df.iloc[:, 4:126]
Y = df.iloc[:, 126]

#Generating 5 different test train splits
Xtrain1, Xtest1, ytrain1, ytest1 = train_test_split(X, Y, test_size = 0.2 , random_state =10)
Xtrain2, Xtest2, ytrain2, ytest2 = train_test_split(X, Y, test_size = 0.2 , random_state =0)
Xtrain3, Xtest3, ytrain3, ytest3 = train_test_split(X, Y, test_size = 0.2 , random_state =100)
Xtrain4, Xtest4, ytrain4, ytest4 = train_test_split(X, Y, test_size = 0.2 , random_state =500)
Xtrain5, Xtest5, ytrain5, ytest5 = train_test_split(X, Y, test_size = 0.2 , random_state =750)

#Putting the five train and test datasets into csv files 
#Training data 1
CandC_train_num1 = Xtrain1
CandC_train_num1  = pd.concat([CandC_train_num1 , ytrain1], axis=1)
CandC_train_num1 = CandC_train_num1.reset_index(drop=True)
ytrain1 = ytrain1.reset_index(drop=True)
CandC_train_num1.to_csv("CandC_train_num1.csv")

#Test Data 1
CandC_test_num1 = Xtest1
CandC_test_num1  = pd.concat([CandC_test_num1 , ytest1], axis=1)
CandC_test_num1 = CandC_test_num1.reset_index(drop=True)
ytest1 = ytest1.reset_index(drop=True)
CandC_test_num1.to_csv("CandC_test_num1.csv")



#Training data 2
CandC_train_num2 = Xtrain2
CandC_train_num2  = pd.concat([CandC_train_num2 , ytrain2], axis=1)
CandC_train_num2 = CandC_train_num2.reset_index(drop=True)
ytrain2 = ytrain2.reset_index(drop=True)
CandC_train_num2.to_csv("CandC_train_num2.csv")

#Test Data 2
CandC_test_num2 = Xtest2
CandC_test_num2  = pd.concat([CandC_test_num2 , ytest2], axis=1)
CandC_test_num2 = CandC_test_num2.reset_index(drop=True)
ytest2 = ytest2.reset_index(drop=True)
CandC_test_num2.to_csv("CandC_test_num2.csv")



#Training data 3
CandC_train_num3 = Xtrain3
CandC_train_num3  = pd.concat([CandC_train_num3 , ytrain3], axis=1)
CandC_train_num3 = CandC_train_num3.reset_index(drop=True)
ytrain3 = ytrain3.reset_index(drop=True)
CandC_train_num3.to_csv("CandC_train_num3.csv")

#Test Data 3
CandC_test_num3 = Xtest3
CandC_test_num3  = pd.concat([CandC_test_num3 , ytest3], axis=1)
CandC_test_num3 = CandC_test_num3.reset_index(drop=True)
ytest3 = ytest3.reset_index(drop=True)
CandC_test_num3.to_csv("CandC_test_num3.csv")



#Training data 4
CandC_train_num4 = Xtrain4
CandC_train_num4  = pd.concat([CandC_train_num4 , ytrain4], axis=1)
CandC_train_num4 = CandC_train_num4.reset_index(drop=True)
ytrain4 = ytrain4.reset_index(drop=True)
CandC_train_num4.to_csv("CandC_train_num4.csv")

#Test Data 4
CandC_test_num4 = Xtest4
CandC_test_num4  = pd.concat([CandC_test_num4 , ytest4], axis=1)
CandC_test_num4 = CandC_test_num4.reset_index(drop=True)
ytest4 = ytest4.reset_index(drop=True)
CandC_test_num4.to_csv("CandC_test_num4.csv")



#Training data 5
CandC_train_num5 = Xtrain5
CandC_train_num5  = pd.concat([CandC_train_num5 , ytrain5], axis=1)
CandC_train_num5 = CandC_train_num5.reset_index(drop=True)
ytrain5 = ytrain5.reset_index(drop=True)
CandC_train_num5.to_csv("CandC_train_num5.csv")

#Test Data 5
CandC_test_num5 = Xtest5
CandC_test_num5  = pd.concat([CandC_test_num5 , ytest5], axis=1)
CandC_test_num5 = CandC_test_num5.reset_index(drop=True)
ytest5 = ytest5.reset_index(drop=True)
CandC_test_num5.to_csv("CandC_test_num5.csv")
```

```python
from sklearn.linear_model import LinearRegression
clf = LinearRegression()

print("\n_______________________________________Linear Regression___________________________________________\n")
print("\n//////////All Coefficients have been stored in .csv files in the same folder////////////////////\n")

#For set 1
clf.fit(Xtrain1, ytrain1)
y_pred1 = clf.predict(Xtest1)
rss1 = np.sum(np.square(ytest1-y_pred1))

Coefficient_LR_set_1 = clf.coef_
pd.DataFrame(Coefficient_LR_set_1).to_csv("Coefficient_LR_set_1.csv")
print("\nFor set 1 rss1 = " ,rss1)


#For set 2
clf.fit(Xtrain2, ytrain2)
y_pred2 = clf.predict(Xtest2)
rss2 = np.sum(np.square(ytest2-y_pred2))
Coefficient_LR_set_2 = clf.coef_
pd.DataFrame(Coefficient_LR_set_2).to_csv("Coefficient_LR_set_2.csv")
print("\nFor set 2 rss2 = " ,rss2)

#For set 3
clf.fit(Xtrain3, ytrain3)
y_pred3 = clf.predict(Xtest3)
rss3 = np.sum(np.square(ytest3-y_pred3))
Coefficient_LR_set_3 = clf.coef_
pd.DataFrame(Coefficient_LR_set_3).to_csv("Coefficient_LR_set_3.csv")
print("\nFor set 3 rss3 = " ,rss3)

#For set 4
clf.fit(Xtrain4, ytrain4)
y_pred4 = clf.predict(Xtest4)
rss4 = np.sum(np.square(ytest4-y_pred4))
Coefficient_LR_set_4 = clf.coef_
pd.DataFrame(Coefficient_LR_set_4).to_csv("Coefficient_LR_set_4.csv")
print("\nFor set 4 rss4 = " ,rss4)

#For set 5
clf.fit(Xtrain5, ytrain5)
y_pred5 = clf.predict(Xtest5)
rss5 = np.sum(np.square(ytest5- y_pred5))
Coefficient_LR_set_5 = clf.coef_
pd.DataFrame(Coefficient_LR_set_5).to_csv("Coefficient_LR_set_5.csv")
print("\nFor set 5 rss5 = " ,rss5)

Average_rss = (rss1+rss2+rss3+rss4+rss5) / 5
print("\nAverage rss of 5 sets ::::::::" ,Average_rss)
```
